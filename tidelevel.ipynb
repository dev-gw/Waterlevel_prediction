```python
# Consider NULL
k = 20220401
df = pd.DataFrame(columns=['tide_level'])
for i in range(30):
  url = 'http://www.khoa.go.kr/api/oceangrid/tideObs/search.do?ServiceKey=9M3GoYlr/kzaB86oY1wO3w==&ObsCode=DT_0032&Date={}&ResultType=json'.format(k)
  print(k)
  response = urllib.request.urlopen(url)
  response_message = response.read().decode('utf8')
  data = json.loads(response_message)
  temp = pd.DataFrame(columns=['tide_level'])
  # empty data case
  try:
    l=data['result']['data']
  except:
    print('=====empty data====')
    for i in range(1440):
      temp.loc[i] = np.nan
    temp = pd.DataFrame(temp[::10])
    df = df.append(temp, ignore_index=True)
    k+=1
    continue
  k_str = str(k)
  datetime_str = k_str[0:4] + '-' + k_str[4:6] + '-' + k_str[6:8] +  ' 00:00:00'
  datetime_obj = datetime.datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S')
  k+=1
  i = 0
  j = 0
  # 1 month -> 1440 minutes
  for _ in range(1440):
    # sequence data not exist
    try:
      comp = l[j]['record_time']
    except:
      j -= 1
      comp = l[j]['record_time']
    comp = comp[:17] + '00'
    comp = datetime.datetime.strptime(comp, '%Y-%m-%d %H:%M:%S')
    if comp == datetime_obj:
        temp.loc[i] = l[j]['tide_level']
        i += 1
        j += 1
    else:
      temp.loc[i] = np.nan
      i += 1
    datetime_obj = datetime_obj + datetime.timedelta(minutes=1)
  print(temp.isna().sum())
  print('\n')
  temp = pd.DataFrame(temp[::10])
  df = df.append(temp, ignore_index=True)
```
